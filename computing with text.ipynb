## Part one:

In this portion of the project, I'm going to display the frequency of the twenty most common words in a bar graph for each text against thier frequencies as the response variable. This will require a few functions to help execute the code properly and efficently.

```python
# Import Around the World in Eighty Days text as a link
ur1 = "https://scassani.github.io/MTH337-S2021/_downloads/58e66eb33ff8fbf7872fbfb57be3f12c/around_world.txt"
world = requests.get(ur1).text

# Import The Republic as a link
ur2 = "https://scassani.github.io/MTH337-S2021/_downloads/eab27d9178a30da5d59f566929d006f3/republic.txt"
republic = requests.get(ur2).text

# Import the Little Women as a link
ur3 ="https://scassani.github.io/MTH337-S2021/_downloads/42ab21da14ec897449a9ad044d08c0c9/little.txt"
little = requests.get(ur3).text

# Import Gulliver’s Travels into Several Remote Regions of the World as a link
ur4 = "https://scassani.github.io/MTH337-S2021/_downloads/74af82a6b009f7e6692546a8bc09cbda/gulliver.txt"
gulliver = requests.get(ur4).text
```

```python
skip = ".,;:()_-\"?!<>$&*#\n\r[]\{\}" # Defines a variable skip that'll allow us to accurately clean the text
```

```python
def cleaned(text): # Defines a function that will allow us to successfully clean the text file
    x=text.lower() # Makes text file lower case
    for char in skip: # Create a loop to eliminate the charachters in the skip variable
        x=x.replace(char,' ') # Replaces each character with a space
    split =x.split(' ')  # Splits the text file at the spaces
    
    book=[] # Define a list called books to store cleaned file
    for i in split: #initialize a loop to store cleaned texts
        if i != '': # adds each characther that is not a space
            book.append(i) # adds each one of said characters to the book list
    
    return book
```

```python
print(cleaned(republic)) # Testing the function to see if it runs correctly by entering the Republic text file
```

```python
def counter(text): # Defines the function counter which counts the number of each word and charachter 
    count = {} # Defines a dictionary to store values and items
    for words in text: # Creates loop to count the words in the text
        if words in count: # If statement to count the words
            count[words]+=1 # If statment adding the characters 
        else: 
            count[words]=1 #If words not in count than the  charachter remains the same    
    return count # Returns the value
```

```python
counter(republic) # Testing to see if the function works using the republic file
```

```python
def sortlist(k): #Defines the function sortlist to start the list with the first element
    return k[1] #returns the first element in the list
```

```python
def sort(counter): ## Defines function to have each value in descending order which makes it 
                    #easy to identify the most common words 
    return sorted(counter,key=sortlist,reverse=True)
```

# Around The World 

```python
cleanw=cleaned(world) # assign a variable to the cleaned around the world text
countw=counter(cleanw) # Assign a variable to the frequency
wlist=list(countw.items()) # Lists both words and frequency
sword=sort(wlist) # Sorts frequency of each word in descending order
print(sword[:20]) #prints list of the twenty most common words.
```

```python
frequencyw=[k[1] for k in sword[:20]] #Takes the frequency of each word
VerneGraph = plt.bar(range(1,21),frequencyw,color='cyan',align='center') #Bar graph for Around the World
plt.title('Around the World in Eighty Days Word Frequency',fontsize=12) # Title
plt.xlabel('Word',fontsize=10) #Xlabel
plt.xticks(np.arange(1,21), sword ,rotation =90) # x-axis Label features
plt.ylabel('Frequency',fontsize=10) # Y-axis label
plt.show(); #Displays graph
```

```python
x1=np.arange(1,len(sword)+1)
y1=np.array([k[1] for k in sword])

x2=np.arange(1,len(rword)+1)
y2=np.array([k[1] for k in rword])

x3=np.arange(1,len(lword)+1)
y3=np.array([k[1] for k in lword])

x4=np.arange(1,len(gword)+1)
y4=np.array([k[1] for k in gword])



def logplot(x,y,t=""):
    
    plt.figure(figsize=(12,6))
    plt.subplot(141)
    plt.plot(x,y)
    plt.title("x ; y")
    plt.subplot(142)
    plt.plot(np.log(x),y)
    plt.title("log-x ; y")
    plt.subplot(143)
    plt.plot(x,np.log(y))
    plt.title("x ; log-y")
    plt.subplot(144)
    plt.plot(np.log(x),np.log(y))
    plt.title("log-x ; log-y")
    
    plt.suptitle(t)

    return None
```

```python
logplot(x1,y1,"Around the World in Eighty Days: \n Graphs of log transformed variables \n")
```

```python
logplot(x2,y2,"The Republic:\n Graphs of Transformed Variables ")
```

```python
logplot(x3,y3,"Little Women:\n Graph of Transformed Variables")
```

```python
logplot(x4,y4," Gulliver’s Travels into Several Remote Regions of the World:\n Graphs of Transformed Variables")
```

```python
#Plots the log(x) vs. log(y) functions for the four texts on the same graph to clearly see correlation
plt.figure(figsize=(8,5))
plt.plot(np.log(x1),np.log(y1),'cyan')
plt.plot(np.log(x2),np.log(y2),'crimson')
plt.plot(np.log(x3),np.log(y3),'palegreen')
plt.plot(np.log(x4),np.log(y4),'yellow')
plt.xlabel(f'$Log(x_i)$',fontsize=16)
plt.ylabel(f'$Log(y_i)$',fontsize=16)
plt.title('$Log(x_i)$ vs. $Log(y_i)$ Frequencies',fontsize=16)
plt.show()
```
