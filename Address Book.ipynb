```python
import requests # Import the neccessary package to carry out the project (allows us to get the web links)
import re # Imports regular expressions package which allows us to look for particular string charachters
```

```python
url = "http://pharmacy.buffalo.edu/faculty-staff/faculty.html" #The Website where the information is coming from for the project
pharm = requests.get(url).text #create variable for the pharamacy faculty to retrieve the information on the website
pharm[:1000] # Calls variable to see if it runs fine and check certain details to ensure the project runs smoothly
```

```python
pharmlink=re.findall(r'"facultyindexname" href="(.*?)">',pharm)#Pharmlink gets each link for each faculty member in the directory 
pharmlink # Call pharmalink to check to see the links for each member
len(pharmlink) # Checks to see the number of faculty memebers in the Pharmacy School based on the number of links
```

```python
# There are 57 links meaning there are 57 faculty members
for i in pharmlink: # Creates a for loop to iterate through each faculty link
    fac=requests.get(i).text # Retrieves the info from each link so the neccessary text can be extracted
    
    #*Name of each faculty member*#
    
    fname=re.findall(r'<span id="lblFacultyName">(.*?)</span>',fac)[0]#Filters the text to find the names of the faculty members
    print(fname) # Prints the resulting filtered text which results in the name of each faculty member
   
    #*Address for each faculty member*#
    
    addy=re.findall(r'"lblAcademicOfficeAddressMobile">(.*?)</a><br/>',fac) #Addy finds the address for each member,addy offers a range of text to select from
    address=re.findall(r'(.*?)Phone:',addy[0]) 
    if len(address)==0: # Checks to see if the address string is equal to zero charachters
        address='No Address' # If equal to zero, python will print no address
    else:
        address=address[0].replace('<br/>',' ') #Otherwise it'll get the actual address of the staff memeber 
    print(address) #Prints the address after being filtered

    #*Phone number of each faculty member*#
    
    number=re.findall(r'Phone: (.*?) <br/>',addy[0]) # filters the addy variable text to retreive the phone number
    if len(number)==0: # if the length of the string is equal to zero 
        number='No Phone Number' # Then Python will then print "No Phone Number" 
    else:
        number=number[0] 
        n=re.findall(r'\d',number) # Reads all the digits in the number variable
        number='('+''.join(n[:3])+') '+ (''.join(n[3:6])) + '-'+(''.join(n[6:10])) # Joins each section of the number string
    print(number) # Prints the resulting phone number

    #*Email for each faculty member*#
    
    email=re.findall(r'href="mailto:(.*?)">',addy[0]) # Filters text to find the email for each faculty member
    if len(email)==0: #If the length of the email is equal to zero then 
        email='No Email' # python will assign no email 
    else:
        email=email[0] #Other wise it'll start with the first elements
    print(email) # Print's the email of the faculty member or prints no email
    print(' ') # Prints a space between each iteration to allow for readability between faculty members 
```
